```python
import mysql.connector

remote = mysql.connector.connect(
    host = "database-1.cpueyaq8kvyb.ap-northeast-2.rds.amazonaws.com",
    port = 3306,
    user = "root",
    password = "960920kmj",
    database = "leaddb"
)

cur = remote.cursor()

try:
    # SQL 파일 읽기
    with open("test04.sql") as code:
        sql = code.read()
    
    # 단일 execute로 다중 쿼리 실행
    cur.execute(sql)
    
    # 결과셋이 있는 경우 처리
    if cur.with_rows:
        results = cur.fetchall()
        print(results)
    
    # 다음 결과셋이 있는지 확인하고 처리
    while cur.nextset():
        if cur.with_rows:
            results = cur.fetchall()
            print(results)
    
    # 모든 결과셋 처리 후 커밋
    remote.commit()
    
except mysql.connector.Error as err:
    print(f"오류: {err}")
    remote.rollback()
    
finally:
    cur.close()
    remote.close()
```

```python
class GetDatabase:

    def __init__(self, host="database-1.cpueyaq8kvyb.ap-northeast-2.rds.amazonaws.com",
                port=3306,
                user="root",
                password="960920kmj",
                database="leaddb"):
        self.host = host
        self.port = port
        self.user = user
        self.password = password
        self.database = database
        self.conn = None
        self.createConnection()

    def createConnection(self):
        self.conn = mysql.connector.connect(
            host = self.host,
            port = self.port,
            user = self.user,
            password = self.password,
            database = self.database
        )
        
    def closeConnection(self):
        self.conn.close()
    
    def processSql(self, sql):
        cursor = self.conn.cursor(buffered=True)
        cursor.execute(sql)
        self.conn.commit()
        return cursor.fetchall()
    
    def processSqlWithParams(self, sql, params):
        cursor = self.conn.cursor(buffered=True)
        cursor.execute(sql, params)
        self.conn.commit()
        return cursor.fetchall()
    
        
    def rollback(self):
        self.conn.rollback()
        
    def __del__(self):
        self.closeConnection()
```

```python
import pandas as pd
df = pd.read_csv("data/gas_station.csv", encoding="utf-8")
df.head()
```

```python
import mysql.connector
from tqdm import tqdm

# GetDatabase 클래스 인스턴스 생성
db = GetDatabase()

# SQL 쿼리 정의
sql = """INSERT INTO crime_status VALUES ("2020", %s, %s, %s, %s)"""

# 데이터프레임의 각 행을 데이터베이스에 삽입
try:
    for i, row in tqdm(df.iterrows(), total=len(df)):
        # processSqlWithParams 메서드 사용
        db.processSqlWithParams(sql, tuple(row))
except Exception as e:
    print(f"오류 발생: {e}")
    db.rollback()  # 오류 발생 시 롤백
finally:
    # 연결 종료 (del 메서드가 자동으로 호출되지만, 명시적으로 호출하는 것이 좋습니다)
    db.closeConnection()
```

```python
#1. 경찰서 별 절도 범죄 평균 발생 건수를 가장 많은 건수 순으로 10개 검색
sql = """select police_station, avg(case_number) as avg_num from crime_status where crime_type like '절도' group by police_station order by avg_num desc limit 10;"""
```

```python
#2. 경찰서 별로 가장 많이 검거한 범죄 건수를 '가장 적은 건수 순'으로 5개 검색
sql = """
SELECT cs.police_station, cs.crime_type, cs.case_number
FROM crime_status cs
JOIN (
    SELECT police_station, MAX(case_number) as max_num
    FROM crime_status
    WHERE status_type = '검거'
    GROUP BY police_station
) sub ON cs.police_station = sub.police_station AND cs.case_number = sub.max_num
ORDER BY cs.case_number ASC
LIMIT 5;"""
```

```python
#3. 경찰서 별로 가장 적게 검거한 건수 중 4건보다 큰 경우를 건수가 큰 순으로 정렬하여 검색
sql = """
SELECT cs.police_station, cs.crime_type, cs.case_number
FROM crime_status cs
JOIN (
    SELECT police_station, MIN(case_number) as min_num
    FROM crime_status
    WHERE status_type = '검거'
    GROUP BY police_station
) sub ON cs.police_station = sub.police_station AND cs.case_number = sub.min_num
WHERE cs.case_number > 4
ORDER BY cs.case_number DESC;"""
```

```python
#4. ‘대문’으로 끝나는 이름의 경찰서 별 범죄발생 건수의 평균이 500건 이상인 경우를 검색
sql = """select police_station, avg(case_number) from crime_status where police_station like '%대문%' group by police_station having avg(case_number) >= 500;"""
```
