```python
import requests
import time
import pandas as pd

url = "https://dapi.kakao.com/v2/local/search/keyword.json"
headers = {"Authorization": "KakaoAK 3c35ad0167f6fe1fb74ac7793ce01549"}

all_places = []
max_pages = 45  # Kakao API 최대 페이지 수
page = 1

while page <= max_pages:
    # '이디야 서울'로만 검색 (rect 파라미터 없음)
    params = {
        'query': '이디야서울',
        'page': page,
        'size': 15  # 최대값
    }
    
    response = requests.get(url, params=params, headers=headers)
    result = response.json()
    
    # 에러 체크
    if 'documents' not in result:
        print(f"에러 발생: {result}")
        break
    
    # 결과가 없으면 종료
    if not result['documents']:
        print(f"페이지 {page}에 결과가 없습니다. 종료합니다.")
        break
        
    # 결과 추가
    documents = result['documents']
    all_places.extend(documents)
    
    # 진행상황 출력
    print(f"페이지 {page}/{max_pages}: {len(documents)}개 결과 수집 (현재까지 총 {len(all_places)}개)")
    
    # 메타데이터 확인 (첫 페이지만)
    if page == 1:
        print(f"총 검색 결과: {result['meta']['total_count']}개")
        print(f"페이지 접근 가능 결과: {result['meta']['pageable_count']}개")
    
    # 마지막 페이지이면 종료
    if result['meta']['is_end']:
        print("마지막 페이지에 도달했습니다.")
        break
        
    # 다음 페이지로
    page += 1
    time.sleep(0.3)  # API 서버 부하 방지를 위한 대기

# 최종 결과 보고
print(f"\n최종 수집 결과: {len(all_places)}개 서울 이디야 매장")
```

```python
import requests
import json
import pandas as pd
import random
url = "https://dapi.kakao.com/v2/local/search/keyword.json"
headers = {"Authorization": "KakaoAK 3c35ad0167f6fe1fb74ac7793ce01549"}
```

```python
def getLocalData(query, page, x=None, y=None, radius=None):
    headers = {"Authorization": "KakaoAK 3c35ad0167f6fe1fb74ac7793ce01549"}

    if x and y and radius:
        params = {'query': query, 'page' : page, 'size' : 15, 'x' : x, 'y' : y, "radius" : radius}
    else:
        print("INFO : 중심점 좌표 입력 안됨")
        params = {'query': query, 'page' : page, 'size' : 15}

    request = requests.get(url, params=params, headers=headers).json()
    return request
```

```python
def parsingStoreInfo(place):
    placeName = place["place_name"]
    placeUrl = place["place_url"]
    placeAddress = place["road_address_name"]
    placeDistance = place["distance"]
    placePhone = place["phone"]
    placeGroupCode = place["category_group_code"]
    placeCategoryName = place["category_name"]
    return (
        placeName,
        placeUrl,
        placeAddress,
        placeDistance,
        placePhone,
        placeGroupCode,
        placeCategoryName
    )
```

```python
def gatheringAroundStoreInfo(query, lastPage, coordinateX, coordinateY, radius):
    storeList = pd.DataFrame(columns=["업체명", "카테고리코드", "카테고리", "전화번호", "홈페이지", "거리"])

    for cnt in range(1, lastPage + 1, 1):
        print(cnt, "페이지 읽는중", end="\r")
        places = getLocalData(query, cnt, coordinateX, coordinateY, radius=700)

        meta = places['meta']
        places = places['documents']

        for place in places:
            (
                placeName,
                placeUrl,
                placeAddress,
                placeDistance,
                placePhone,
                placeGroupCode,
                placeCategoryName
            ) = parsingStoreInfo(place)
            storeList.loc[len(storeList)] = [
                placeName,
                placeGroupCode,
                placeCategoryName,
                placePhone,
                placeUrl,
                placeDistance
            ]
    
        if meta['is_end']:
            print("마지막 페이지 입니다.")
            break
        if cnt == lastPage:
            print("최대 페이지 한도 도달!")
            
    return storeList
```

```python
lastPage = 10

places = getLocalData("대륭테크노타운3차", 1)
coordinateX, coordinateY = places['documents'][0]['x'], places['documents'][0]['y']
print("좌표 : ", coordinateX, coordinateY)

storeList = gatheringAroundStoreInfo("가산디지털단지맛집", 10, coordinateX, coordinateY, 700)

storeList["카테고리"] = storeList["카테고리"].apply(lambda x : x.split(" > ")[1].strip())

print("검색된 상점의 카테고리에는", storeList["카테고리"].unique(), "가 있습니다.")
preferCategory = input("이중 특별히 원하시는 카테고리가 있으신가요? 없으면 엔터")

if preferCategory:
    idxList = storeList[storeList["카테고리"] == preferCategory].index
    storeInfo = storeList.loc[random.choice(idxList)]
else:
    storeInfo = storeList.loc[random.choice(storeList.index)]

print(f"오늘 {storeInfo['업체명']}는 어떠세요?")
print(f"카테고리는 {storeInfo['카테고리']}에 속하고, 거리는 {storeInfo['거리']}m 떨어져 있습니다.")
print(f"전화번호는 {storeInfo['전화번호']}이고, 홈페이지는 {storeInfo['홈페이지']}입니다.")
```
