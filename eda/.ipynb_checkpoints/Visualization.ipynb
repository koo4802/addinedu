## 시각화 조금 더 효과적으로

### 평범한 그리기

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
import numpy as np

x= np.linspace(0, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 3*np.sin(x+1.0)
y4 = 4*np.sin(x+1.5)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y1,color='red')
plt.plot(x,y2,color='blue')
plt.plot(x,y3,color='green')
plt.plot(x,y4,color='purple')
plt.show()
```

### 스타일 적용

```python
sns.set_style('dark')
plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
sns.set_style('white')
plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
sns.set_style('whitegrid')
plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
sns.set_style('white')
plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
sns.despine(offset=10)
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.head(5)
```

### 박스플롯

```python
plt.figure(figsize=(8,6))
sns.boxplot(x=tips["total_bill"])
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

### 컬러맵

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips, hue="smoker", palette="Set3")
plt.show()
```

### 스왑플롯

```python
plt.figure(figsize=(8,6))
sns.swarmplot(x="day", y="total_bill", data=tips, color='0.5')
plt.show()
```

### 박스플롯과 스왑플롯

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, color='0.25')
plt.show()
```

### lm 플롯

```python
sns.set_style('darkgrid')
sns.lmplot(x="total_bill", y="tip", data=tips, height=7)
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, height=7)
plt.show()
```

### flights 데이터셋

```python
flights = sns.load_dataset('flights')
flights.head(5)
```

### pivot 메서드

```python
flights = flights.pivot(index="month", columns="year", values="passengers")
```

```python
flights.head(5)
```

### 히트맵

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt='d')
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt='d', cmap='YlGnBu')
plt.show()


```

### iris 데이터셋

```python
sns.set(style='ticks')
iris = sns.load_dataset('iris')
iris.head(10)
```

```python
sns.pairplot(iris, hue="species")
plt.show()
```

```python
sns.pairplot(iris, x_vars=["sepal_width", "sepal_length"],
            y_vars=["petal_width", "petal_length"],
            hue="species",
            height=4)
plt.show()

```

### titanic 데이터셋, catplot


```python
titanic = sns.load_dataset('titanic')
titanic.head(5)
```

```python
sns.catplot(x="sex", y="survived", hue="pclass", data=titanic, kind="bar",
            palette="muted", legend=True, height=7)
plt.show()
```

```python
sns.catplot(x="deck", kind="count", palette="ch:.25",data=titanic, legend=False)
```
