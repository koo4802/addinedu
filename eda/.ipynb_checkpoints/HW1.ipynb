```python
from numpy.polynomial import Polynomial
f1 = Polynomial.fit(data_result["인구수"], data_result["총계"], 1)

fx = np.linspace(100000, 700000, 100)

data_result["오차"] = data_result["총계"] - f1(data_result["인구수"])

df_descending = data_result.sort_values(by="오차", ascending=False)
df_ascending = data_result.sort_values(by="오차", ascending=True)
```

```python
df_descending.head()
```

```python
from matplotlib.colors import ListedColormap

color_step = ["#e74c3c", "#f1c40f", "#95a5a6","#2ecc71", "#3498db", "#3498db"]
my_colormap = ListedColormap(color_step)
```

```python
def draw_graph9():
    plt.figure(figsize=(14, 10))
    plt.scatter(data_result["인구수"], data_result["총계"], c=data_result["오차"],
                 s=50, cmap=my_colormap)
    plt.plot(fx, f1(fx), ls="dashed", lw=3, color="gray")
    
    for n in range(5):
        plt.text(df_descending["인구수"][n] * 1.02,
                 df_descending["총계"][n] * 0.98,
                 df_descending.index[n],
                 fontsize=15)
        plt.text(df_ascending["인구수"][n] * 1.02,
                 df_ascending["총계"][n] * 0.98,
                 df_ascending.index[n],
                 fontsize=15)
    plt.xlabel("인구수")
    plt.ylabel("CCTV")
    plt.grid()
    plt.show()
                 
draw_graph9()
```

```python
for i in range(9):
    for j in range(9):
        print(f"{i+1}*{j+1}={(i+1)*(j+1)}")
```

```python
for i in range(9):
    for j in range(9):
        print(f"{j+1}*{i+1}={(i+1)*(j+1)}",end ="\t")
    print("")
```

```python
floor = int(input("층수를 입력하세요: "))

# 트리 부분과 기둥 부분의 공통 로직을 함수로 추출
def print_line(spaces, stars):
    print(" " * spaces + "*" * stars + " " * spaces)

# 트리 부분 출력
for i in range(1, floor + 1):
    print_line(floor - i, i * 2 - 1)

# 기둥 부분 출력 - 한 번의 계산으로 재사용
trunk_spaces = floor - 2
for _ in range(3):
    print_line(trunk_spaces, 3)
```

## 데이터 합치기

```python
left = pd.DataFrame(
    {
        "key" : ["K0", "K4", "K2", "K3"],
        "A" : ["A0", "A1", "A2", "A3"],
        "B" : ["B0", "B1", "B2", "B3"]
    }
)

right = pd.DataFrame(
    {
        "key" : ["K0", "K1", "K2", "K3"],
        "C" : ["C0", "C1", "C2", "C3"],
        "D" : ["D0", "D1", "D2", "D3"]
    }
)

```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on="key")
```

```python
pd.merge(left, right, on="key", how="left")
```

```python
pd.merge(left, right, on="key", how="right")
```

```python
pd.merge(left, right, on="key", how="outer")
```

```python
pd.merge(left, right, on="key", how="inner")
```

```python
import numpy as np

t = np.arange(start=0, stop=12, step=0.01)
y = np.sin(t)
```

```python
def draw_graph():
    plt.figure(figsize=(10, 6))
    plt.plot(t, np.sin(t), label="sin")
    plt.plot(t, np.cos(t), label="cos")
    plt.grid()
    plt.legend()
    plt.xlabel("time")
    plt.ylabel("Amplitude")
    plt.title("Example of sinwave")
    plt.show()

draw_graph()
```

```python
t = np.arange(start=0, stop=5, step=0.5)

def draw_graph2():
    plt.figure(figsize=(10, 6))
    plt.plot(t, t, "r--")
    plt.plot(t, t**2, "bs")
    plt.plot(t, t**3, "g^")
    plt.show()

draw_graph2()
```

```python
t = [0,1,2,3,4,5,6]
y = [1,4,5,8,9,5,3]

def draw_graph3():
    plt.figure(figsize=(10, 6))
    plt.plot(t, y, color="green", linestyle="dashed",
             marker="o", markerfacecolor="blue", markersize=12)
    plt.xlim([-0.5, 6.5])
    plt.ylim([0.5, 9.5])
    plt.show()

draw_graph3()
```
