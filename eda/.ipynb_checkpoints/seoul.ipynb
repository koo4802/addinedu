```python
# 3x3 서브플롯 생성
f, ax = plt.subplots(3, 3, figsize=(20, 20))

# 데이터와 제목 매핑
titles = {
    '스트레스': '전체 학생 스트레스',
    '스트레스남학생': '남학생 스트레스',
    '스트레스여학생': '여학생 스트레스',
    '우울감경험률': '전체 학생 우울감',
    '우울남학생': '남학생 우울감',
    '우울여학생': '여학생 우울감',
    '자살생각율': '전체 학생 자살생각',
    '자살남학생': '남학생 자살생각',
    '자살여학생': '여학생 자살생각'
}

# 각 서브플롯에 파이 차트 그리기
for idx, col in enumerate(col_names):
    row = idx // 3  # 행 인덱스
    col_idx = idx % 3  # 열 인덱스
    
    raw_data[col].plot.pie(
        explode=[0, 0.02],
        ax=ax[row, col_idx],
        autopct='%.1f%%',
        title=titles[col]
    )
    ax[row, col_idx].set_ylabel('')

plt.tight_layout()  # 서브플롯 간격 자동 조정
plt.show()
```
