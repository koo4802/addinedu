```python
from bs4 import BeautifulSoup
import pandas as pd
import urllib
from urllib.request import urlopen, Request
from user_agent import generate_user_agent
season_list = []
episode_list = []
title_list = []
story_list = []
rating_list = []
date_list = []
url = "https://www.imdb.com/title/tt0898266/episodes/?season={sinarioNo}&ref_=ttep"
for i in range(1, 13):
    req = Request(url.format(sinarioNo = urllib.parse.quote(str(i))), headers={'User-Agent' : generate_user_agent()})
    page = urlopen(req)
    soup = BeautifulSoup(page, "html.parser")
    
    for each in soup.find_all("div", class_="ipc-html-content-inner-div"):
        story_list.append(each.get_text())

    for idx, each in enumerate(soup.find_all("div", class_="ipc-title__text")):
        episode_num_tem, title_tem = each.get_text().split(" ∙ ")
        season_tem, episode_tem = episode_num_tem.split(".")
        season_list.append(season_tem)
        episode_list.append(episode_tem)
        title_list.append(title_tem)
    for each in soup.find_all("span", class_="ipc-rating-star--rating"):
        rating_list.append(each.get_text())
    for each in soup.find_all("span", class_="sc-f2169d65-10 bYaARM"):
        date_list.append(each.get_text())
    

df = pd.DataFrame({ "season": season_list, "episode": episode_list, "title": title_list, "story": story_list, "rating": rating_list, "date": date_list})
df

```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By


def get_storyline(url):
    driver = webdriver.Chrome()
    try:
        driver.get(url)
        driver.execute_script("window.scrollTo(0, 2000)")
        driver.implicitly_wait(5)
        storyline = driver.find_element(By.CSS_SELECTOR, '[data-testid="Storyline"]').find_element(By.CSS_SELECTOR, 'div.ipc-html-content-inner-div')
        return storyline.text
    finally:
        driver.quit()

from bs4 import BeautifulSoup
import pandas as pd
import urllib
from urllib.request import urlopen, Request
from user_agent import generate_user_agent
season_list = []
episode_list = []
title_list = []
story_list = []
rating_list = []
date_list = []
url = "https://www.imdb.com/title/tt0898266/episodes/?season={sinarioNo}&ref_=ttep"
for i in range(1, 13):
    req = Request(url.format(sinarioNo = urllib.parse.quote(str(i))), headers={'User-Agent' : generate_user_agent()})
    page = urlopen(req)
    soup = BeautifulSoup(page, "html.parser")
    
    for idx, each in enumerate(soup.find_all("div", class_="ipc-title__text")):
        detail_url = "https://www.imdb.com" + each.parent["href"]
        storyline = get_storyline(detail_url)
        story_list.append(storyline)
        
        episode_num_tem, title_tem = each.get_text().split(" ∙ ")
        season_tem, episode_tem = episode_num_tem.split(".")
        season_list.append(season_tem)
        episode_list.append(episode_tem)
        title_list.append(title_tem)
    for each in soup.find_all("span", class_="ipc-rating-star--rating"):
        rating_list.append(each.get_text())
    for each in soup.find_all("span", class_="sc-f2169d65-10 bYaARM"):
        date_list.append(each.get_text())
    

df = pd.DataFrame({ "season": season_list, "episode": episode_list, "title": title_list, "story": story_list, "rating": rating_list, "date": date_list})
df

```
