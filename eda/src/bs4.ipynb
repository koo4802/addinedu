```python
from bs4 import BeautifulSoup
import pandas as pd
from urllib.request import urlopen, Request
from user_agent import generate_user_agent
```

```python
import time
url = "https://search.naver.com/search.naver?query={search_words1}"
req = Request(url.format(search_words1 = urllib.parse.quote("박스오피스")), headers={'User-Agent' :generate_user_agent()})
page = urlopen(req)
soup = BeautifulSoup(page, "html.parser")
```

```python
rating = soup.find(class_= "_panel_rating _tab_content")
movie_list = []
for idx, each in enumerate(rating.find_all("li")):
    title = each.find("strong").string
    rate = each.find("span", class_="sub_text").get_text()
    if "." in title:
        movie_url = "https://search.naver.com/search.naver" + each.find("a")["href"]
        movie_page = urlopen(movie_url)
        movie_soup = BeautifulSoup(movie_page, "html.parser")
        title = movie_soup.find("strong", class_= "_text").get_text()
    print(f"[{idx+1}위 : {title}] {rate}")
    movie_list.append([title, rate, "https://search.naver.com/search.naver" + each.find("a")["href"]])
    time.sleep(0.5)
```

```python
import pandas as pd

df = pd.DataFrame(movie_list, columns=["name", "rating", "link"])
df
```
